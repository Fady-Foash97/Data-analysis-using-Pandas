## Technological products dataset:
- This dataset contains information about various electronic devices from different brands, including details such as the brand name, device type, model, country of origin, release date, and price

## Dataset columns: 
- Rank: The ranking of the device.
- Brand: The manufacturer of the device.
- Device: The category of the device (e.g., smartphone, laptop, console).
- Model: The specific model name or number.
- Country of Origin: The country where the device originates from.
- Date of Release: The release date of the device.
- Price (USD): The price of the device in U.S. dollars.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_excel("Technological products.xlsx")
```

## Goal 1: Identify the most expensive and least expensive devices.

```python
df.info()
```

```python
df
```

```python
df1 = df[["Device", "Brand", "Model", "Price (USD)"]].sort_values(by = "Price (USD)", ascending= False)
```

```python
df1
```

```python
plt.figure(figsize=(10, 12))
plt.barh(df1["Device"], df1["Price (USD)"], color = "green")
plt.xlabel("Price")
plt.ylabel("Device")
plt.title("Price of devices")
plt.show()
```

```python
fig = px.bar(df1, x = "Price (USD)", y = "Device", orientation = "h", color = "Price (USD)", color_continuous_scale = "blues", width = 1000, height = 900)
fig.update_layout(xaxis_title = "Price", yaxis_title = "Device", title = "Price of devices")
fig.show()
```

# Conclusions:
- Dell Alienware and Apple Macbook are the most expensive devices, they cost about 1999 dollars.
- Google nest and Sony WH are the least expensive devices, they cost about 229 and 349 dollars respectively.

## Goal 2: Find the average price of devices by brand

```python
df.info()
```

```python
df2 = df.groupby("Brand")["Price (USD)"].mean().reset_index()
```

```python
df2
```

```python
df3 = df2.sort_values(by = "Price (USD)", ascending = False)
```

```python
df3
```

```python
plt.figure(figsize=(10, 10))
plt.plot(df2["Brand"], df2["Price (USD)"])
plt.xlabel("Brand", size= 15)
plt.ylabel("Price of devices", size = 13)
plt.grid()
plt.title("Average price of devices by brand")
plt.show()
```

```python
fig = px.line(df2, x = "Brand", y = "Price (USD)", width = 1000, height = 900)
fig.update_layout(xaxis_title = "Brand", yaxis_title = "Price of devices", 
                  xaxis = dict(showgrid = True, gridcolor = "lightgrey", gridwidth = 1),
                  yaxis = dict(showgrid = True, gridcolor = "lightgrey", gridwidth = 1),
                  plot_bgcolor = "white",
                  title = "Average price of devices by brand")
fig.show()
```

## Conclusions: 
- Apple is the brand with the highest average price.
- Google is the brand with the lowest average price.
- Apple is famous for two things: Brand loyalty and technological advancements, in 2007: Steve Jobs changed the world with introducing the Apple Iphone, A mobile device that combines the features of: a mobile phone, a camera and a computer, since this event, Apple has updated this device and also made many advanced devices like the macbook and tablets.
- Google is known around the world for its search engine not for its devices, in short: It's a software company not a hardware company like Apple, Dell and Samsung.

## Goal 4: Analyze price trends over time.

```python
df.info()
```

```python
df6 = df.groupby("Year")["Price (USD)"].mean().reset_index()
```

```python
df7 = df6.sort_values(by = "Price (USD)", ascending = False)
```

```python
df7
```

```python
plt.plot(df7["Year"], df7["Price (USD)"])
plt.xlabel("Year")
plt.ylabel("Price trend")
plt.grid()
plt.title("Price trends over time")
plt.show()
```

## Conclusions:
- As we see from the chart, Prices have increased from 2020 (year of Covid 19) to 2021.
- This shows that in 2020, due to worldwide pandemic, people couldn't go to work, therefore no production, which results in no buying and selling.
- but in 2021, the pandemic has stopped and people now can go to work so they can produce something, which results in people buying and selling which results in an active economy.

## Goal 5: Compare brands based on price and number of products.

```python
df.info()
```

```python
Avg_price = df.groupby("Brand")["Price (USD)"].mean().reset_index()
```

```python
Avg_price
```

```python
Product_count = df.groupby("Brand")["Device"].count().reset_index()
```

```python
Product_count
```

```python
plt.figure(figsize=(18, 14))
## First subplot for average price:
plt.subplot(2, 2, 1)
plt.bar(Avg_price["Brand"], Avg_price["Price (USD)"], color = "purple")
plt.xlabel("Brand")
plt.ylabel("Price")
plt.xticks(rotation= 45)
plt.title("Average prices for brand")
plt.show()
plt.figure(figsize=(21, 15))
## First subplot for number of products per brand:
plt.subplot(2, 2, 2)
plt.plot(Product_count["Brand"], Product_count["Device"], color="green")
plt.xlabel("Brand")
plt.ylabel("Device")
plt.title("Number of devices per brand")
plt.grid()
plt.show()
```

## Conclusions:
- The number of devices per brand remains consistent (3 devices per brand).
- But Apple has the highest market share in the tech industry due to its: technological advancements and brand loyalty.
- Next to Apple: HP, Lenovo, Samsung, Asus and Dell.

## Goal 7: Determine the most common device type

```python
df.info()
```

```python
df[["Device", "Model", "Type"]]
```

```python
device_type = df.groupby("Type")["Device"].count().reset_index()
```

```python
device_type
```

```python
device_type_arrange = device_type.sort_values(by = "Device", ascending = False)
```

```python
device_type_arrange
```

```python
plt.figure(figsize=(12, 10))
plt.barh(device_type_arrange["Type"], device_type_arrange["Device"])
plt.xlabel("Number of devices")
plt.ylabel("Type of device")
plt.title("Number of devices per type")
plt.show()
```

```python
fig = px.bar(device_type_arrange, x = device_type_arrange["Device"], y = device_type_arrange["Type"], orientation = "h", width = 1000, height = 900)
fig.update_layout(xaxis_title = "Number of devices", yaxis_title = "Type of device", title = "Number of devices per type")
fig.show()
```

## Conclusions: 
- Laptops are the most common device with 18 units, this is due to a variety of laptops' functions: work, study, computing, programming and so on.
- Mobile phones and Ipads are common by 3 or 4 units respectively, they are used for portability and usage.
- Consoles and Pc are common by 2 units respectively, Desktop gaming and gaming consoles have the same level of demand.
- Headphones are the least common by only one unit, that means there's less demand for this hardware.
