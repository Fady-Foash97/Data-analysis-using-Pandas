## Pixar movies dataset:

- This dataset contains financial, critical, and audience reception data for Pixar movies. It includes 28 entries, each representing a different Pixar film, along with multiple attributes detailing their performance and reception. 

## dataset columns:

- ID ‚Äì A unique identifier for each film.
- Film ‚Äì The title of the Pixar movie.
- Film Rating ‚Äì The movie‚Äôs MPAA rating (e.g., G, PG).
- Cinema Score ‚Äì The audience score from CinemaScore, representing general public reception (e.g., A+, A, A-).
- Release Date ‚Äì The date when the movie was released in theaters.
- Budget ‚Äì The estimated production cost of the film in dollars.
- Box Office US & Canada ‚Äì Total revenue generated from ticket sales in the United States and Canada.
- Box Office Other ‚Äì Total revenue from international markets (outside the US & Canada).
- Box Office Worldwide ‚Äì The sum of US & Canada and international box office earnings.
- Rotten Tomatoes Score ‚Äì The critic approval rating (percentage) from Rotten Tomatoes.
- IMDb Score ‚Äì The average audience rating of the film on IMDb (on a scale of 1 to 10).

2. Audience Reception & Ratings ‚≠ê
- Identify which films have the highest Rotten Tomatoes and IMDb scores.

## Q3: Find the ROI (Return on Investment) for each film: ROI= Box Office Worldwide - Budget / Budget * 100

```python
df.head()
```

```python
df["ROI (Return on Investment)"] = df["box_office_worldwide"] - df["budget"] / df["budget"] * 100
```

```python
df.info()
```

```python
df_sorted = df[["ROI (Return on Investment)", "film"]].sort_values(by = "ROI (Return on Investment)", ascending = False)
```

```python
fig = px.bar(df, x = df_sorted["film"], y = df_sorted["ROI (Return on Investment)"], width= 1200, height = 900)
fig.update_layout(xaxis_title= "Film", yaxis_title = "Return on investment", title = "Return on investment for film")
fig.show()
```

## Conclusion: 

- Inside out 2 has the highest return on investment, that means it's the most profitable movie in Pixar's catalogue.
- Reasons are Strong audience reception and controlled production costs.
- Several other films have strong ROI: Incredibles 2, Toy story 4, Finding Dory, Inside out, Toy story 3, Finding Nemo.
- These films benefited from strong brand recongition (Toy story, Incredibles and Finding Nemo)
- Films with the lowest ROI are: Onward, Soul, Luca, Turning red, Lightyear, Elemental
- Reasons are:
- Luca, Turning red and Soul got Disney plus releases before theaters.
- Onward was released in March 2020 before worldwide pandemic shut down.
- Lightyear had low audience reception due to unethical marketing techniques, generic plot, no emotional core and poor pacing.

## Goal 3 : Release Date Trends:

## Q1: Analyze which decades had the most successful releases.

```python
df.info()
```

```python
df[["film", "release_date"]]
```

```python
df.info()
```

```python
df["release_date"] = df["release_date"].astype("datetime64[ns]")
```

```python
df.info()
```

```python
df["decade"] = (df["release_date"].dt.year // 10 * 10).astype(str) + "s"
```

```python
df.info()
```

```python
df_decade = df.groupby("decade")["film"].count().reset_index()
```

```python
df_decade
```

```python
fig = px.bar(df_decade, x = df_decade["film"],  y = df_decade["decade"], 
             color = df.groupby("decade")["Critical reception"].mean(), 
             color_continuous_scale = "blues")
fig.update_layout(xaxis_title = "Number of films", yaxis_title = "Decade" ,title = "Most successful decade in Pixar")
fig.show()
```

## Conclusions:
- The 1990s and the 2000s are the most successful decades for Pixar:
- - Hit after hit: Toy Story, Toy Story 2, A bug's life, Monsters inc, Incredibles and many more!
- The 2010s is the start of the dark age for pixar:
- - Although there were hits like Inside out, Toy story 3 and Coco.
  - There were some flops like incredibles 2, The good dinosaur, Finding Dory, Cars 2 and Cars 3.
- The 2020s is the decade where Pixar struggled, although there were a few hits like Soul and Inside out 2, there were several challenges:
- - The launch of the Disney Plus platform.
  - Theaters having weak receptions due to the pandemic.
  - Weak receptions for Lightyear, Turning red and Elemnetal due to corporate decisions that Disney (the owner of Pixar) made.

## Goal 4: Franchise Success
## Compare Toy Story films vs. other Pixar franchises.

```python
df["Franchise"] = df["film"].apply(lambda x: "Toy Story" if "Toy Story" in x else "Other Pixar")
```

```python
df
```

```python
df_Franchise = df.groupby("Franchise")["film"].count().reset_index()
```

```python
df_Franchise
```

```python
fig = px.bar(df_Franchise, x =  df_Franchise["Franchise"], y =  df_Franchise["film"]  , color = df.groupby("Franchise")["Critical reception"].mean() ,  
             color_continuous_scale = "blues")
fig.update_layout(xaxis_title = "Franchise", yaxis_title = "Number of films")
fig.show()
```

## Conclusions: 
- Although the toy story franchise has about 4 films which is lower than Pixar's massive catalogue, it's still the most successful Pixar franchise until now.
- Pixar focuses on making standalone movies rather than franchise movies like Dreamworks (Kung fu Panda, Shrek, How to train you dragon), Disney (Lion king, Mulan, Wreck it ralph) and Sony (Spiderverse and Cloudy).

## Analyze if sequels tend to perform better or worse than original movies.

```python
df["Sequel or original"] = df["film"].apply(lambda x: "Sequel" if any(n in x for n in ["2", "3", "4"])  else "Original")
```

```python
df
```

```python
fig = px.bar(df, x = df["film"], y = df["Critical reception"], 
             color = df["Sequel or original"],
             color_discrete_map ={"Original": "green", "Sequel": "Purple"}, 
             width = 1200, height = 900)
fig.update_layout(xaxis_title = "film", yaxis_title = "Critical reception", title = "Originals and sequels success rate", 
                  xaxis = dict(tickangle = -35))
fig.show()
```

## Conclusions:

- A lot of Pixar's most successful movies are original movies with the exception of Cars, Brave, the good dinosaur and Elemental.
- Some Pixar sequels are hits like:
- - The toy story sequels
- - Incredibles 2 
- - Inside out 2
- Cars is the least popular Pixar franchise in terms of critical reception.
