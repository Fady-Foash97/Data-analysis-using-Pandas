## Columns:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandasql as psql
```

```python
df = pd.read_csv('ai_job_dataset.csv')
```

```python
df.drop(columns = ['job_description_length'], inplace = True)
```

```python
df.head()
```

## Goal 1: Which job title that has the highest salary? 

```python
Job_title = """SELECT job_title, AVG(salary_usd) AS Salary FROM df
GROUP BY job_title"""
```

```python
Job_title1 = psql.sqldf(Job_title, globals())
```

```python
Job_title1
```

```python
Job_title_sort = """SELECT job_title, Salary FROM Job_title1
ORDER BY Salary DESC"""
```

```python
Job_title_sort1 = psql.sqldf(Job_title_sort, globals())
```

```python
Job_title_sort1
```

```python
plt.figure(figsize = (10, 13))
sns.lineplot(data = Job_title_sort1, x = Job_title_sort1['Salary'], y = Job_title_sort1['job_title'], marker = 'o')
plt.xlabel('Salary', size = 12)
plt.ylabel('Job', size = 12)
plt.title('Salary per Job', size = 12)
plt.tight_layout()
plt.show()
```

## Conclusions:
- Jobs related to AI(AI specialist, Machine learning engineer, Head of AI, AI research scientist, AI architect, Robotics engineer) are the highest-paid jobs. 
- Alternatively, Data analyst is the lowest paid job despite its importance in decision making and its signficant position in the job market.
- This reflects the position and rise of AI as an emergent field since the creation of CHAT GPT.

## Goal 5: Relationship between years of experience and experience_level?

```python
Experience_level_min = df.groupby('experience_level')['years_experience'].min().reset_index()
```

```python
Experience_level_min
```

```python
Experience_level_min_sort = Experience_level_min.sort_values(by = 'years_experience', ascending = True)
```

```python
Experience_level_min_sort
```

```python
Experience_level_max = df.groupby('experience_level')['years_experience'].max().reset_index()
```

```python
Experience_level_max_sort = Experience_level_max.sort_values(by = 'years_experience', ascending = True)
```

```python
Experience_level_max_sort
```

```python
plt.figure(figsize = (12, 10))
sns.lineplot(data = Experience_level_min_sort, x = Experience_level_min_sort['experience_level'], y = Experience_level_min_sort['years_experience'], 
            label = 'Minimum year', color = 'green')
sns.lineplot(data = Experience_level_max_sort, x = Experience_level_max_sort['experience_level'], y = Experience_level_max_sort['years_experience'],
            label = 'Maximum year')
plt.xlabel('Experience level', size = 12)
plt.grid()
plt.ylabel('Years of experience', size = 12)
plt.title('Experience level vs Years of experience', size = 12)
plt.show()
```

## Conclusions:
- There's a positive relationship between the experience level and years of experience:
- Entry level jobs require 0 to 1 year experience.
- Mid level jobs require 2 to 4 years.
- Senior level jobs require 5 to 9 years.
- Executive level jobs require 10 to 19 years.

## Goal 6: What are the most required skills for AI jobs?

```python
Required_skills = df.explode('required_skills')['required_skills'].value_counts()
```

```python
Required_skills = Required_skills.groupby(Required_skills.index.str.strip()).sum().sort_values(ascending = False)
```

```python
Required_skills
```

```python
plt.figure(figsize = (13, 10))
plt.pie(Required_skills[:10].values, labels = Required_skills.index[:10], autopct = '%1.1f%%', startangle = 90)
plt.title('Top 10 Most required skills for AI jobs')
plt.axis('equal')
plt.show()
```

```python
Bottom_skills = Required_skills.tail(10)
plt.figure(figsize = (13, 10))
plt.pie(Bottom_skills[:10].values, labels = Bottom_skills.index[:10], autopct = '%1.1f%%', startangle = 90)
plt.title('10 least required skills for AI jobs')
plt.axis('equal')
plt.show()
```

## Goal 7: Remote jobs vs non remote jobs?

```python
df['remote_status'] = df['remote_ratio'].apply(lambda x: 'remote' if x == 0 else 'Not remote')
```

```python
df
```

```python
df.info()
```

```python
Remote = """SELECT remote_status, job_title FROM df"""
```

```python
Remote_1 = psql.sqldf(Remote, globals())
```

```python
Remote_1
```

```python
Remote_2 = """SELECT remote_status, job_title, COUNT(*) AS Count From Remote_1
GROUP BY remote_status, job_title
ORDER BY Count DESC"""
```

```python
Remote_3 = psql.sqldf(Remote_2)
```

```python
Remote_3
```

```python
plt.figure(figsize= (10 , 12))
sns.barplot(data = Remote_3, x = Remote_3['Count'], y = Remote_3['remote_status'], hue = 'job_title', palette = 'Purples')
plt.xlabel('Job count')
plt.ylabel('Jobs')
plt.tight_layout()
plt.title('Remote vs Not remote jobs')
plt.show()
```

# Conclusions:
- There are a lot of non remote job openings than remote job openings.

## Goal 9: Relation between company size and remote ration.

```python
df
```

```python
Company_remote = """SELECT company_size, remote_ratio, Count(*) AS Count FROM df
GROUP BY company_size, remote_ratio
ORDER BY remote_ratio"""
```

```python
Company_remote1 = psql.sqldf(Company_remote)
```

```python
Company_remote1
```

```python
plt.figure(figsize = (14 , 12))
sns.barplot(data = Company_remote1, x = Company_remote1['company_size'], y = Company_remote1['Count'], hue = 'remote_ratio')
plt.legend(loc = "lower left")
plt.xlabel('Size of company')
plt.ylabel('Remote ratio')
plt.tight_layout()
plt.title('Relation between size of company and remote ratio')
plt.show()
```

## Conclusions:
- Small enterprises(S) have high job counts across all remote ratios showing flexibility for job hirings in start ups.
- Medium-sized companies (M) show a slightly lower count for fully remote roles, possibly favoring hybrid (50%) or onsite work.
- Large companies (L) have fewer fully remote (100) jobs compared to their onsite (0) and hybrid (50%) roles, indicating they prefer employees to be on-premises or partially remote.
- Startups and small companies are more open to remote work, while larger organizations lean toward in-office or hybrid models, likely due to infrastructure, security, and collaboration preferences.
