## Bestseller Books dataset:
- Bestseller Book Dataset provides the data about bestselling books, author name, reviews, user rating, price, year and genre of so many books.

## Columns:
- Name: name of the book
- Author: name of the author of the book
- User Rating: rating given by user to the book
- Reviews: reviews of the book
- Price: Price of the given book
- year: publishing year of the book
- Genre: genre of the given book

## Goal 6: User rating vs Price (Do high prices mean high user rating?)

```python
User_rating_vs_Price = df.groupby('User Rating')['Price'].mean().round().reset_index()
```

```python
User_rating_vs_Price
```

```python
User_rating_vs_Price_sort = User_rating_vs_Price.sort_values(by = 'Price', ascending = False)
```

```python
User_rating_vs_Price_sort
```

```python
plt.figure(figsize = (10, 12))
sns.barplot(data = User_rating_vs_Price_sort, y = User_rating_vs_Price_sort['Price'], 
            x = User_rating_vs_Price_sort['User Rating'], color = 'orange')
plt.xlabel('User rating', size = 17)
##plt.xticks(rotation = 45, ha = 'right')
plt.ylabel('Price', size = 17)
plt.tight_layout()
plt.title('Price vs user rating')
plt.show()
```

# Conclusions:
- There is a non-linear relationship between user rating and price, high prices do not gurantee high user rating and vice versa.
- Some of the highest rated books have moderate prices, which shows that Readers value content over cost.
- Some expensive books such as Academic studies and box sets have good ratings, likely due to their format or collector appeal especially on box sets.
- Lower rated books are expensive, these books could be either polarizing books or disposable quick reads.
- Overall: what drives reader satisfaction is not the value or cost of the book but the content underneath, its relevance and timelessness.
