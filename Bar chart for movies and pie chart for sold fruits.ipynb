```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.array(["Avatar", "Avengers Endgame", "Avatar the way of water", "Titanic", "Star wars the force awakens", "Avengers Infinity War", "Spider-Man No Way Home", "Inside Out 2", "Jurassic World", "The Lion King", "The Avengers", "Furious 7", "Top Gun: Maverick"])
y = np.array([2923706026, 2797501328, 2320250281, 2257844554, 2068223624, 2048359754, 1922598800, 1698863816, 1671537444, 1656943394, 1518815515, 1515341399, 1495696292])
plt.barh(x, y, height=0.4, color="green")
plt.xlabel("Box office")
plt.title("Highest grossing movies")
plt.show()
```
