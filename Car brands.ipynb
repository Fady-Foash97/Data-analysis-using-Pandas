## Large Dataset of Cars:

- This dataset provides a comprehensive overview of various vehicles and their key technical specifications.
- It includes detailed attributes such as brand, model, vehicle class, region of origin, drivetrain type, engine size, horsepower, and fuel efficiency.

## Columns:

- Brand – car brand
- Model – specific model name of the car
- VehicleClass – body type
- Region – geographical region where the car was manufactured
- DriveTrain – type of drivetrain system
- MSRP – manufacturer’s suggested retail price
- DealerCost – the price at which the car is sold by the dealer
- EngineSize – size of the engine in liters
- Cylinders – number of cylinders in the engine
- HorsePower – maximum engine power output in horsepower
- MPG_City – fuel economy in miles per gallon while driving in city conditions
- MPG_Highway – fuel economy in miles per gallon while driving on highways
- Weight – weight of the vehicle in pounds
- Wheelbase – distance between the front and rear wheels
- Length – total length of the vehicle from bumper to bumper

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandasql as psql
```

## Goal 1: Which brand has the highest value?

```python
df["Value"] = df["MSRP"] - df["DealerCost"]
```

```python
df
```

```python
Brand = df.groupby('Brand')['Value'].mean().reset_index()
```

```python
Brand
```

```python
Sorted_brand = Brand.sort_values(by = "Value", ascending = False)
```

```python
Sorted_brand
```

```python
plt.figure(figsize = (10, 12))
sns.barplot(data = Sorted_brand, x = Sorted_brand['Value'], y = Sorted_brand['Brand'], color = 'Green')
plt.xlabel("Value", size = 12)
plt.ylabel("Brand", size = 12)
plt.title("Brand by value" , size = 12)
plt.tight_layout()
plt.show()
```

## Goal 3: Car type Weight vs value

```python
Query = """SELECT VehicleClass, COUNT(Weight) as Weight, AVG(Value) as Average_value
FROM df
GROUP BY VehicleClass"""
```

```python
Weight_vs_value = psql.sqldf(Query)
```

```python
Weight_vs_value
```

```python
Sorted_value = Weight_vs_value.sort_values(by = "Average_value", ascending = False)
```

```python
Sorted_value
```

```python
Sorted_class = Weight_vs_value.sort_values(by = "Weight", ascending = False)
```

```python
Sorted_class
```

```python
plt.figure(figsize = (10, 12))
sns.barplot(data = Sorted_class, x = Sorted_class['VehicleClass'], y = Sorted_class['Weight'], hue = "Average_value")
plt.xlabel("VehicleClass", size = 12)
plt.ylabel("Weight", size = 12)
plt.title("Weight of car class based on average value" , size = 12)
plt.show()
```

## Conclusions:
- There is no positive relation between a car type's weight and average value.
- Car types' average value do not increase according to weight.
- Sports and SUV vehicle types have lower weight than sedan vehicle type.
- Although the sedan vehicle type has the highest average value.
- The main factor is luxury: Both Sports cars and SUVS offer luxurious value, while sedans, wagons, trucks and hybrid vehicles are suitable for every day activities.
## In short:
- A vehicle type's average value depends on luxury more than utility.
