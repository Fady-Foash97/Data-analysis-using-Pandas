## Instagram data dataset:

- Instagram generates a lot of data every day. When content creators post on Instagram, they do look at their engagement and reach to find the category of posts they should post more.

## Columns:

- Impressions: Total number of times the post was viewed.
- From Home: Number of views coming from the home feed.
- From Hashtags: Number of views coming from hashtag searches.
- From Explore: Number of views coming from the Explore page.
- From Other: Views from other sources not specified.
- Saves: Number of times the post was saved by users.
- Comments: Total number of comments on the post.
- Shares: Number of times the post was shared.
- Likes: Total number of likes received.
- Profile Visits: Number of users who visited the profile from this post.
- Follows: Number of new followers gained from this post.
- Caption: Text description or message included in the post.
- Hashtags: List of hashtags used in the post.

# What data that needs to be analyzed?

- What is the post that has the highest impression or reach? does higher impression means higher engagement?
- What sources that bring the most visits?
- What sources that bring the most follows?
- What hashtags that drive the most views and engagements?

# Mission 1: What is the post that has the highest impression or reach? does higher impression means higher engagement?

```python
df["Engagement"] = df["Likes"] + df["Shares"] + df["Comments"] + df["Saves"]
```

```python
df.info()
```

```python
df[["Post number","Likes", "Comments", "Shares", "Saves" , "Engagement", "Impressions"]].sort_values(by="Impressions", ascending = False)
```

```python
plt.figure(figsize= (10,10))
plt.scatter(df["Engagement"], df["Impressions"])
plt.grid()
plt.xlabel("Engagement")
plt.ylabel("Impressions")
plt.title("Engagement vs Impressions")
plt.show()
```

## Conclusions:

## Posts with high impressions or reach does not mean high Engagement

- As we can see from the following graph and the table above:
- post number 119  has the highest impressions but less engagements.
- (36919 impressions but 1127 engagements)
- Post number 118 has the highest engagement but less impressions than post number 119, most of these engagements come from the likes users make.
- (1721 engagements but 32695 impressions)

## Decisions:

- We should make posts and stories that provide our users with incentives to engage with us like:
- stories and posts that have links to click on to visit our homepage on the web (if we have one).
- Clips of our courses.
- Opinions of our students.
- Visually engaging content.

## Mission 3: What sources that bring the most follows?

```python
df[["Profile Visits", "Follows" , "From Home", "From Hashtags", "From Explore", "From Other"]].sort_values(by="Follows", ascending= False)
```

## Conclusions:

- Not all profile visits are converted to follows.
- A lot of follows comes from hashtags.

## Decisions:

## Mission 4: What hashtags that drive the most views and engagements?

```python
df.info()
```

```python
df
```

```python
df["Hashtags"] = df["Hashtags"].apply(lambda x: str(x).split())
```

```python
Repeated_Hashtag = df.explode("Hashtags")["Hashtags"].value_counts()
```

```python
Repeated_Hashtag
```

```python
plt.figure(figsize= (10,10))
plt.pie(Repeated_Hashtag[:10], labels= Repeated_Hashtag.index[:10])
plt.legend(loc="upper left")
plt.title("Top 10 repeated hashtags")
plt.show()
```

```python
plt.figure(figsize= (12,6))
plt.bar(Repeated_Hashtag.index[:10], Repeated_Hashtag.values[:10])
plt.xlabel("Hashtag")
plt.ylabel("Repeated")
plt.xticks(rotation= 30, ha="right")
plt.title("Top 10 repeated hashtags")
plt.show()
```

## Conclusions:

- The 10 most repeated hashtags in the instagram profile are: <b>#thecleverprogrammer, #amankharwal, #python, #machinelearning, #pythonprogramming, #datascience, #ai, #pythonprojects, #artificalintelligence</b> and <b>#data</b>

- These hashtags have something in common: <ul>they are related to trends concerning AI and its advancements in the last few years especially in the field of programming and data analytics, people are very interested in this topic.</u>

# Decisions:

- We should focus on writing hashtags related to: AI, programming and data analytics, even slight varitations of them so we can bring more traffic to our instagram profile
